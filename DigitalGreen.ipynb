```python
import pandas as pd

train_data = pd.read_csv(r"C:\Users\Dennis\Desktop\Machine learning\Digital Green Crop Yield Estimate Challenge\Train.csv")
test_data= pd.read_csv(r"C:\Users\Dennis\Desktop\Machine learning\Digital Green Crop Yield Estimate Challenge\Test.csv")
train_data
```

```python
import pandas as pd

# Assuming your DataFrame is named 'df'

# List of column names to be converted into categorical
categorical_columns = [
    'District', 'Block', 'LandPreparationMethod', 'CropEstMethod', 
    'SeedingSowingTransplanting', 'NursDetFactor', 'TransDetFactor', 
    'TransplantingIrrigationSource', 'TransplantingIrrigationPowerSource',
    'PCropSolidOrgFertAppMethod', 'CropbasalFerts', 'MineralFertAppMethod', 
    'FirstTopDressFert', 'MineralFertAppMethod.1', 'Harv_method',
    'Threshing_method', 'Stubble_use'
]

# Convert selected columns to categorical data type
train_data[categorical_columns] = train_data[categorical_columns].astype("category")
```

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score

X = train_data.drop(['ID', 'Yield'], axis=1)
X = X.select_dtypes(include=np.number)
y = train_data['Yield']

# Split the dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=1234)

# Model Selection and Training (Linear Regression)
model = LinearRegression()
model.fit(X_train.fillna(0), y_train)

# Model Evaluation
y_pred = model.predict(X_test.fillna(0))
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

# Print the evaluation metrics
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2): {r2}")
```

```python
test_df = test_data[X.columns]
preds = model.predict(test_df.fillna(0))
test_data
```
